```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset/payment_history.csv')
df.info()
```

```python
df
```

```python
fraud_mapping = {False:0, True:1}
df['is_fraud'] = df['is_fraud'].map(fraud_mapping)
df
```

```python
status_mapping = {'Pending':1, 'Failed':2, 'Success': 3}
df['payment_status'] = df['payment_status'].map(status_mapping)
df
```

```python
y = df.is_fraud
```

```python
payment_features = ['payer_account_number', 'payment_amount', 'payment_limit', 'payment_status']
X = df[payment_features]
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor

payment_model = DecisionTreeRegressor(random_state=1)

payment_model.fit(train_X, train_y)
```

```python
val_predictions = payment_model.predict(val_X)
val_predictions
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(val_y, val_predictions)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(val_y, val_predictions)
accuracy
```
